- **LIBRERIAS.**

```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

- **FUNCIONES.**

```python
nltk.download('vader_lexicon')
sia = SentimentIntensityAnalyzer()
def analyze_sentiment(text):
    sentiment_scores = sia.polarity_scores(text)
    compound_score = sentiment_scores['compound']
    if compound_score < -0.05:
        return 0  # Malo
    elif compound_score > 0.05:
        return 2  # Positivo
    else:
        return 1  # Neutral
```

```python
from joblib import Parallel, delayed

# Definir la función de análisis de sentimiento
def analyze_sentiment(text):
    # Verificar si el texto es un valor float
    if isinstance(text, float):
        return 1  # Se puede asignar un valor neutral por defecto o manejarlo de acuerdo a tus necesidades
    else:
        sentiment_scores = sia.polarity_scores(text)
        compound_score = sentiment_scores['compound']
        if compound_score < -0.05:
            return 0  # Malo
        elif compound_score > 0.05:
            return 2  # Positivo
        else:
            return 1  # Neutral

# Determinar el número de núcleos de CPU disponibles
num_cores = 20  # Reemplaza <Número_de_núcleos> con el número de núcleos que deseas utilizar

# Aplicar análisis de sentimiento en paralelo
sentiments = Parallel(n_jobs=num_cores)(delayed(analyze_sentiment)(text) for text in estados['text'])

# Agregar los resultados al DataFrame
estados['sentiment_analysis'] = sentiments


```
