```python
# instanciamos ambos datasets

yelp = pd.read_csv('abfss://558991ad-6fdf-463a-b58a-238ddabaa398@onelake.dfs.fabric.microsoft.com/37a0382d-f783-4728-a697-48b1dc002937/Files/DATA MODELO/Yelp_Modelo.csv')
google = pd.read_csv('abfss://558991ad-6fdf-463a-b58a-238ddabaa398@onelake.dfs.fabric.microsoft.com/37a0382d-f783-4728-a697-48b1dc002937/Files/DATA MODELO/Google_Modelo.csv')
```

```python
# Definir las condiciones
condicion_1 = (df_total1['RATING_HOTEL'] > 5.1) & (df_total1['RATING_HOTEL'] <= 10)
condicion_2 = (df_total1['RATING_HOTEL'] > 10) & (df_total1['RATING_HOTEL'] <= 15)
condicion_3 = (df_total1['RATING_HOTEL'] > 15)

# Aplicar las condiciones y dividir los valores correspondientes
df_total1.loc[condicion_1, 'RATING_HOTEL'] /= 2
df_total1.loc[condicion_2, 'RATING_HOTEL'] /= 3
df_total1.loc[condicion_3, 'RATING_HOTEL'] /= 4
```
